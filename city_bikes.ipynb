```python
#Get an API for my selected city - Paris
import requests
import json
import pandas as pd
import numpy as np
import json
from pandas import json_normalize
response_paris = requests.get('http://api.citybik.es/v2/networks/velib')    
```

```python
# Check if the request was successful (a status code 200 should be given)
if response_paris.status_code != 200:
    print("Request failed. Status code:", response_paris.status_code)
else:
    # Parse the JSON response
    data = response_paris.json()
    
    # Extract bike stations in Paris
    stations = data['network']['stations']
    
    # Create lists for extracted details
    name = []
    latitudes = []
    longitudes = []
    number_of_bikes = []
    
    # Iterate through each station and extract details
    for station in stations:
        name.append(station['name'])
        latitudes.append(station['latitude'])
        longitudes.append(station['longitude'])
        number_of_bikes.append(station['free_bikes'])  
```

```python
# Create a DataFrame with the extracted details
df_paris = pd.DataFrame({
    'Station Name': name,
    'Latitude': latitudes,
    'Longitude': longitudes,
    'Number of Bikes': number_of_bikes
})   
```
