# Foursquare

Send a request to Foursquare with a small radius (1000m) for all the bike stations in your city of choice. 

Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

```python
df2_barcelona = pd.read_csv("../Downloads/barcelona_bikes.csv")     

results = [] 

for index, row in df2_barcelona.iterrows():
    latitude = row['Latitude']
    longitude = row['Longitude']


    url = "https://api.foursquare.com/v3/places/search"

    params = {
        'll': f'{latitude},{longitude}',
        'categoryId': '13000,12013',    # for restaurants and bars
        'radius': 1000,
        'fields': 'name,location,rating,categories,distance,description,popularity,price',
        'sort': 'RATING'
    }

    headers = {
        "Accept": "application/json",
        "Authorization": "fsq3m5MbEYLkUCci/RnQ/24+FkYWsR79yM/2FgAV3+VN7ec="
    }

    response = requests.request("GET", url, params=params, headers=headers)
    if response.status_code != 200:
        print("Request failed. Status code:", response.status_code)
    else:
    # Parse the JSON response
        data = response.json()       

     
```

Put your parsed results into a DataFrame

```python
res = data.get('results', [])
fr = pd.DataFrame(res)   
results.append({
    'Latitude': latitude,
    'Longitude': longitude,
    'Foursq. Name': fr['name'],
    'Distance': fr['distance'],
    'Rating': fr['rating'],
    'Price': fr['price'],
    'Popularity': fr['popularity']
    })                    
     
```

```python
data_foursquare = pd.DataFrame(results)

data_foursquare.head(1)              
```

```python
# Specify the file path for the CSV
csv_file_path = "../Downloads/foursquare_places.csv"

# Save the DataFrame to CSV
data_foursquare.to_csv(csv_file_path, index=False)            
```

```python
# Transforming the data
data_foursquare['Average_dist_F'] = data_foursquare['Distance'].apply(lambda x: sum(x) / len(x))
data_foursquare['Average_popularity_F'] = data_foursquare['Popularity'].apply(lambda x: sum(x) / len(x))
data_foursquare['Average_rating_F'] = data_foursquare['Rating'].apply(lambda x: sum(x) / len(x))

data_foursquare['Max_rating_F'] = data_foursquare['Rating'].apply(lambda x: max(x))
data_foursquare['Max_popularity_F'] = data_foursquare['Popularity'].apply(lambda x: max(x))

data_foursquare['Min_rating_F'] = data_foursquare['Rating'].apply(lambda x: min(x))
data_foursquare['Min_popularity_F'] = data_foursquare['Popularity'].apply(lambda x: min(x))         

 
```
